```python
df=df.drop(['home_team_continent', 'away_team_continent', 'city','country', 
            'shoot_out', 'home_team_mean_midfield_score', 'home_team_mean_offense_score', 
            'home_team_goalkeeper_score', 'home_team_mean_defense_score', 
            'away_team_mean_midfield_score', 'away_team_mean_offense_score', 
            'away_team_goalkeeper_score',  'away_team_mean_defense_score'], axis=1)
```

```python
label_encoder=LabelEncoder()
```

```python
for col in df.columns:
    df[col] = label_encoder.fit_transform(df[col])
    df[col] = df[col].astype('float32')
```

```python
X=df.drop(["home_team_result"],axis=1)
```

```python
y=df["home_team_result"]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train=np.array(X_train)
y_train=np.array(y_train)
X_test=np.array(X_test)
y_test=np.array(y_test)
```

```python
X_train=X_train.reshape((16027,11,1))
X_test=X_test.reshape((7894,11,1))
```

```python
y_train=y_train.reshape((16027,1))
y_test=y_test.reshape((7894,1))
```

```python
model=tf.keras.models.Sequential([
    tf.keras.layers.LSTM(128, activation='relu'),
    tf.keras.layers.LSTM(64, activation='relu'),
    tf.keras.layers.LSTM(32, activation='relu'),
    tf.keras.layers.LSTM(16, activation='relu'),
    tf.keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer=tf.keras.optimizers.legacy.Adam(),loss=tf.keras.losses.binary_crossentropy,metrics=['accuracy'])
```

```python
model.fit(x=X_train,y=y_train)
```
